#### Q1. Find the number of episodes?

```python
shark_tank["episode_number"].nunique()
```

#### Q2. Max, min , mean of asked amount,asked equity, asked valuation?

```python
shark_tank[["pitcher_ask_amount","ask_equity","ask_valuation"]].agg({"max","min","mean"})
```

#### Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise? 

```python
shark_tank.groupby("episode_number")[["pitcher_ask_amount","ask_equity","ask_valuation"]].agg({"max","min"})
```

#### Q4. Brand names in which 2 ,3 or 4 sharks are invested?

```python
shark_tank[shark_tank["total_sharks_invested"]>1][["brand_name","total_sharks_invested"]].reset_index()
```

#### Q5. Episode wise minimum and maximum of sharks invested?

```python
shark_tank.groupby("episode_number")["total_sharks_invested"].agg({'min','max'})
```

#### Q6. Minimum and Maximum sharks invested in each episode?

```python
shark_tank.groupby("episode_number")["amount_per_shark"].agg({"min","max"})
```

#### Q7. Brand names who ask for 1 crore and got a deal? #### 

```python
shark_tank[shark_tank["deal_amount"]==100][["brand_name","deal_amount"]].reset_index().drop('index',axis=1)
```

#### Q8. Brand names where deal equity is between 0 to 50? 

```python
shark_tank[(shark_tank["deal_equity"]>0) & (shark_tank["deal_equity"]<50)][["brand_name","deal_equity"]]
```

#### Q9. Find the number of brands participated in each episode?

```python
shark_tank.groupby('episode_number')["brand_name"].count()
```

#### Q10. How many sharks participated in this show and What were their names? #### 

```python
sharks=shark_tank.columns[shark_tank.columns.str.contains('present')==True].str.split('_')
```

```python
sharks_name=[]
numberOfSharks=0;
for shark in sharks:
    sharks_name.append(shark[0])
    numberOfSharks+=1
    
print(f"Name of the Sharks:{sharks_name}")
print(f"Number of Sharks:{numbersOfSharks}")
```
